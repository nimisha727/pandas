# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending per student actually (\$645-675) underperformed compared to schools with smaller budgets (<\$585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name","school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_schools = len(school_data_complete["school_name"].unique())
total_students = len(school_data_complete["student_name"])
total_budget = school_data["budget"].sum()
totalav_math_score = round(school_data_complete["math_score"].mean(), 2)
totalav_read_score = round(school_data_complete["reading_score"].mean(),2)
overall_pass_rate = round((totalav_math_score + totalav_read_score)/2,2)

math_70 = school_data_complete.loc[school_data_complete["math_score"] >= 70]
math_70_count = len(math_70["math_score"])
pass_math_70 = round(((math_70_count*100)/total_students),2)
read_70 = school_data_complete.loc[school_data_complete["reading_score"] >= 70]
read_70_count = len(read_70["reading_score"])
pass_read_70 = round(((read_70_count*100)/total_students),2)


                
total_schools
total_students
total_budget
totalav_math_score
totalav_read_score
pass_read_70
pass_math_70
overall_pass_rate

school_summary = pd.DataFrame({"TotalSchools": [str(total_schools)],
                              "Total Students":[str(total_students)],
                              "Total Budget":["$ " + str(total_budget)],
                              "Avg Math Score":[str(totalav_math_score)],
                               "Avg Read Score":[str(totalav_read_score)],
                               "% Passing Math":[str(pass_math_70)],
                               "% Passing Read":[str(pass_read_70)],
                               "% Overall Passrate":[str(overall_pass_rate)]
                                })
school_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
schools = school_data_complete.set_index("school_name").groupby("school_name")
sch_type = school_data.set_index("school_name")["type"]
sch_budget = school_data.set_index("school_name")["budget"]
stu_per_sch = schools["Student ID"].count()
per_stu_budget = round(school_data.set_index("school_name")["budget"]/school_data.set_index("school_name")["size"],2)
av_math_score = schools["math_score"].mean()
av_reading_score = schools["reading_score"].mean()
pass_math = round((school_data_complete[school_data_complete["math_score"] >= 70].groupby("school_name")["Student ID"].count()/stu_per_sch*100),2)
pass_reading = round((school_data_complete[school_data_complete["reading_score"] >= 70].groupby("school_name")["Student ID"].count()/stu_per_sch*100),2)
overall_passrate = round((school_data_complete[(school_data_complete["reading_score"] >=70) & (school_data_complete["math_score"] >=70)].groupby("school_name")["Student ID"].count()/stu_per_sch*100),2)

```

```python

sch_summary = pd.DataFrame({
    "School Type": sch_type,
    "Total Students": stu_per_sch,
    "Total School Budget": sch_budget,
    "Per Student Budget": per_stu_budget,
    "Average Math Score": av_math_score,
    "Average Reading Score": av_reading_score,
    "% Passing Math": pass_math,
    "% Passing Reading": pass_reading,
    "% Overall Passing Rate": overall_passrate})   
sch_summary.head()      
```

```python
Top_sch_performance = sch_summary.sort_values("% Overall Passing Rate",ascending=False)

Top_sch_performance.head(5)
```
